```python
import pandas as pd
```

```python
data=pd.read_csv("nba.csv")
```

```python
data
```

```python
### str.lower() this function converts all uppercase characters to lowercase.
```

```python
data["Name"]=data["Name"].str.upper()
```

```python
data
```

```python
data["College"]=data["College"].str.lower()
```

```python
data
```

```python
### Splitting and Replacing a Data
```

```python
data.dropna(inplace=True)
```

```python
data["College"] = data["College"].str.split("a", n = 1, expand=True)
```

```python
 data
```

```python
data["Number"]=data["Number"].replace(99.0,"ninty nine")
```

```python
data
```
